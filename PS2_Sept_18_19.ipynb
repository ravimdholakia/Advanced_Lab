```python
#Decimal to binary conversion

import numpy as np

DecimalInput=np.int64(input("Input decimal number to be converted: "))

binary=np.zeros(50)

def ConverttoBinary(DecimalInput):
    for i in range(0,50):
        if((DecimalInput%2)>0):
            binary[i]=10**i
            DecimalInput=(DecimalInput-(DecimalInput%2))/2
        else:
            DecimalInput=DecimalInput/2
    return(sum(binary))

print(DecimalInput, " is ", int(ConverttoBinary(DecimalInput)), " in binary.")
```

```python
#Newman 3.7

import numpy as np
import matplotlib.pyplot as plt

def Mandelbrot(Resolution, x_length, y_length):
    x_complex = np.linspace(-2, 2, x_length)
    y_complex = np.linspace(-2, 2, y_length)

    complex_constant = x_complex[:,np.newaxis] + 1j*y_complex[np.newaxis,:]

    #Iteration process:

    z = complex_constant
    for j in range(Resolution):
        z = z**2 + complex_constant

    Mandelbrot_set = (abs(z) < 2)

    return Mandelbrot_set

Mandelbrot_set = Mandelbrot(1000,200,200)

plt.plot(np.transpose(Mandelbrot_set))
plt.title("Mandelbrot Distribution")
plt.show()

plt.imshow(Mandelbrot_set.T,'Blues')
plt.title("Mandelbrot Fractal")
plt.show()
```
