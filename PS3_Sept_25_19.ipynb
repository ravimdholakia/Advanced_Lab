```python
#Problem 1: 4.3 in Newman

import numpy as np
import math
import matplotlib.pyplot as plt

#f(x) = x(x-1) = x^2-x
#derivative = 2x-1

y=0

def Func1(x):
    y=x*(x-1)
    return y

#value=input("put in a value: ")

#print(Func1(int(value)))

def derivative(delta):
    derivative_at_point=(Func1(1+delta)-Func1(1))/delta
    return derivative_at_point

print("Derivative when delta=10^-2: ",derivative(0.01))

print("Derivative when delta=10^-4: ",derivative(0.0001))
print("Derivative when delta=10^-6: ",derivative(0.000001))
print("Derivative when delta=10^-8: ",derivative(0.00000001))
print("Derivative when delta=10^-10:",derivative(0.0000000001))
print("Derivative when delta=10^-12:",derivative(0.000000000001))
print("Derivative when delta=10^-14:",derivative(0.00000000000001))
```

```python
#Problem 2

import numpy as np
import math
import matplotlib.pyplot as plt

N1=10
N2=30
N3=100

C1 = np.zeros([N1,N1],float)
C2 = np.zeros([N2,N2],float)
C3 = np.zeros([N3,N3],float)

A1 = np.random.rand(N1,N1)
B1 = np.random.rand(N1,N1)

A2 = np.random.rand(N2,N2)
B2 = np.random.rand(N2,N2)

A3 = np.random.rand(N3,N3)
B3 = np.random.rand(N3,N3)


for i in range(N1-1):
    for j in range(N1-1):
        for k in range(N1-1):
            C1[i,j] += A1[i,k]*B1[k,j]

for i in range(N2-1):
    for j in range(N2-1):
        for k in range(N2-1):
            C2[i,j] += A2[i,k]*B2[k,j]

for i in range(N3-1):
    for j in range(N3-1):
        for k in range(N3-1):
            C3[i,j] += A3[i,k]*B3[k,j]
            
print("Explicit method" ,N1,"x",N1,": ",C1)
print("Explicit method" ,N2,"x",N2,": ",C2)
print("Explicit method" ,N3,"x",N3,": ",C3)
print("np.dot() method" ,N1,"x",N1,": ",np.dot(A1,B1))
print("np.dot() method" ,N2,"x",N2,": ",np.dot(A2,B2))
print("np.dot() method" ,N3,"x",N3,": ",np.dot(A3,B3))
```

```python
#Problem 3
#10,000 1-D random walks of 1,000 steps each

import numpy as np
import math
import matplotlib.pyplot as plt

N_steps=1000
N_walks=10000

x=np.linspace(0,N_steps-1,N_steps)
y=np.zeros(N_steps)

plt.plot(x,y,'r-')
colors=("b-","c-","k-","r-","g-","m-","y-","gold-")

sigma_squared=1

avg=np.zeros(N_steps)

for i in range(1,N_walks):
    color_choice=colors[np.random.randint(0,7)]
    for i in range(1,len(x)):
        y[i]=y[i-1]+(sigma_squared*np.random.randn(1))
        avg[i]=np.mean(y)
    plt.plot(x,y,color_choice)
    
#plt.plot(x,y,'r-')

plt.title("Random Walks")
plt.xlabel("Steps")
plt.ylabel("Distance from 0")
plt.axhline(color="black")
plt.axvline(color="black")
plt.show()

plt.plot(x,avg,'b.')
plt.title("Average by step")
plt.xlabel("Steps")
plt.ylabel("Average")
plt.show()
```
